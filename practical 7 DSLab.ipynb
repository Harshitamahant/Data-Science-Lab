```python
import pandas as np
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import skew, kurtosis
iris = sns.load_dataset('iris')
series = iris['sepal_length']
mean_val = series.mean()
median_val = series.median()
std_val = series.std()
variance_val = series.var()
skewness_val = skew(series)
kurtosis_val = kurtosis(series)
print("Descriptive Statistical Analysis (Iris - Sepal Length)")
print("-" * 50)
print(f'Mean: {mean_val:.2f}')
print(f'Median: {median_val:.2f}')
print(f'Standard Deviation: {std_val:.2f}')
print(f'Variance: {variance_val:.2f}')
print(f'Skewness: {skewness_val:.2f}')
print(f'Kurtosis: {kurtosis_val:.2f}')
plt.figure(figsize=(8, 6))
sns.histplot(series, bins=20, kde=True)
plt.title('Distribution of Iris Sepal Length')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Frequency')
plt.show()
```
