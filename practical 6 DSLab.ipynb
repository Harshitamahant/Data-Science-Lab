```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
df = sns.load_dataset('flights')
print(df.head())
```

UNIVARIATE ***VISUALIZATIONS***

```python
sns.histplot(df['passengers'], bins=20, kde=True)
plt.title('Distribution of Monthly Passenger Counts')
plt.xlabel('Passengers')
plt.ylabel('Frequency')
plt.show()
```

```python
sns.boxplot(x='month', y='passengers', data=df)
plt.title('Passenger Count by Month')
plt.xlabel('Month')
plt.ylabel('Passengers')
plt.show()
```

```python
sns.countplot(x='year', data=df)
plt.title('Record Count by Year')
plt.xlabel('Year')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

BIVARIATE **VISUALIZATIONS**


```python
# Pivot table for line plot
pivot = df.pivot_table(index="month", columns="year", values="passengers")
months_order = ['January', 'February', 'March', 'April', 'May', 'June',
                'July', 'August', 'September', 'October', 'November', 'December']
pivot = pivot.reindex(months_order)

pivot.T.plot(figsize=(8, 4), legend=False)
plt.title('Monthly Airline Passengers (1949 - 1960)')
plt.xlabel('Year')
plt.ylabel('Passengers')
plt.show()
```

```python
correlation_matrix = df.corr(numeric_only=True)
sns.heatmap(correlation_matrix, annot=True, cmap='YlGnBu')
plt.title('Correlation Matrix of Airline Passenger Data')
plt.show()
```

```python

yearly = df.groupby('year')['passengers'].sum().reset_index()

sns.scatterplot(x='year', y='passengers', data=yearly)
plt.title('Total Passengers per Year')
plt.xlabel('Year')
plt.ylabel('Total Passengers')
plt.show()
```
